```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is already loaded into a DataFrame named 'df'
df = pd.read_csv('/Users/peterparker/Downloads/archive-2/indexProcessed.csv')

# Select rows where column 'A' equals 'foo'
df = df[df['IndexName'] == 'NYA']
```

```python
df.head()
```

```python
# Ensure the 'Date' column is a proper datetime object and set it as the index
# (If you already did this in the preprocessing step, you can skip these two lines)
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)

# Set the overall size of the plot (width, height) in inches
plt.figure(figsize=(14, 7))

# Plot the 'Adj Close' price
# We use Adj Close as it accounts for dividends and stock splits, providing a truer historical picture
plt.plot(df.index, df['Adj Close'], label='Adj Close Price', color='#1f77b4', linewidth=1.5)

# Add titles and axis labels
plt.title('Historical Index Price Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price', fontsize=12)

# Add a grid to make it easier to track price levels across the chart
plt.grid(True, linestyle='--', alpha=0.6)

# Add a legend
plt.legend(loc='upper left', fontsize=12)

# Automatically adjust subplot parameters to give the plot clear margins
plt.tight_layout()

# Render the plot
plt.show()
```

```python


# Assuming your data is loaded into 'df' and 'Date' is set as the index
# Example: 
# df['Date'] = pd.to_datetime(df['Date'])
# df.set_index('Date', inplace=True)

# Calculate the 50-day Simple Moving Average
df['SMA_50'] = df['Adj Close'].rolling(window=50).mean()

# 1. Set up the figure and subplots
# We create 2 rows and 1 column. 
# 'gridspec_kw' makes the top price chart 3 times taller than the bottom volume chart.
# 'sharex=True' ensures both charts align perfectly on the same dates.
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), 
                               gridspec_kw={'height_ratios': [3, 1]}, 
                               sharex=True)

# ---------------------------------------------------------
# 2. Top Subplot: Price and Moving Average
# ---------------------------------------------------------
ax1.plot(df.index, df['Adj Close'], label='Adj Close Price', color='#1f77b4', linewidth=1.5, alpha=0.8)
ax1.plot(df.index, df['SMA_50'], label='50-Day SMA', color='#ff7f0e', linewidth=2)

ax1.set_title('Historical Index Price and Trading Volume', fontsize=16, fontweight='bold')
ax1.set_ylabel('Price', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend(loc='upper left', fontsize=12)

# ---------------------------------------------------------
# 3. Bottom Subplot: Trading Volume
# ---------------------------------------------------------
# Volume is traditionally represented as a bar chart
ax2.bar(df.index, df['Volume'], color='#7f7f7f', alpha=0.7, width=1.5)

ax2.set_xlabel('Date', fontsize=12)
ax2.set_ylabel('Volume', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.6)

# 4. Final Formatting
# Adjust the layout so the subplots don't overlap
plt.tight_layout()

# Render the combined plot
plt.show()
```
